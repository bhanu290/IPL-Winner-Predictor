```python
def match_progress(x_df, match_id, pipe):
    match = x_df[x_df['id'] == match_id]
    match = match[(match['ball'] == 6)]
    winner = match['winner'].values[0]
    chase = match['batting_team'].values[0]
    temp_df = match[['batting_team', 'bowling_team', 'city', 'target', 'runs_left', 'balls_left', 'wickets_left', 'crr', 'rrr']].dropna()
    temp_df = temp_df[temp_df['balls_left'] != 0]
    result = pipe.predict_proba(temp_df)
    temp_df['lose'] = np.round(result.T[0]*100,1)
    temp_df['win'] = np.round(result.T[1]*100,1)
    temp_df['end_of_over'] = range(1,temp_df.shape[0]+1)
    target = temp_df['target'].values[0]
    runs = list(temp_df['runs_left'].values)
    new_runs = runs[:]
    runs.insert(0,target)
    temp_df['runs_in_over'] = np.array(runs)[:-1] - np.array(new_runs)
    wickets = list(temp_df['wickets_left'])
    new_wickets = wickets[:]
    new_wickets.insert(0,10)
    temp_df['wickets_in_over'] = np.array(new_wickets)[:-1] -  np.array(wickets)

    print("Target-",target)
    print("Chasing team- ", chase)
    print('Winner- ', winner)
    temp_df = temp_df[['end_of_over','runs_in_over','wickets_in_over','runs_left', 'balls_left','lose','win']]
    return temp_df
```

```python
abbreviation = {
        'csk':'Chennai Super Kings',
        'dc':'Delhi Capitals',
        'mi':'Mumbai Indians',
        'rcb':'Royal Challengers Bangalore',
        'kxip':'Kings XI Punjab',
        'rr':'Rajasthan Royals',
        'srh':'Sunrisers Hyderabad',
        'kkr':'Kolkata Knight Riders'
}
```

```python
def df(batting_team, bowling_team, city, target, runs_left, balls_left, wickets_left, crr, rrr):
    data = [[batting_team, bowling_team, city, target, runs_left, balls_left, wickets_left, crr, rrr]]
    x = pd.DataFrame(data, columns = ['batting_team', 'bowling_team', 'city', 'target', 'runs_left', 'balls_left', 'wickets_left', 'crr', 'rrr'])
    return x
```

```python
batting_team = input("Enter batting team: ").lower()
bowling_team = input("Enter bowling team: ").lower()
city = input("Enter city: ")
target = int(input("Enter target: "))
runs_left = int(input("Enter runs remaining: "))
balls_left = int(input("Enter balls remaining: "))
wickets_left = int(input("Enter wickets remaining: "))
crr = float((target - runs_left)*6/(120 - balls_left))
rrr = float(runs_left*6/balls_left)
x = df(abbreviation[batting_team], abbreviation[bowling_team], city, target, runs_left, balls_left, wickets_left, crr, rrr)
lose = pipe.predict_proba(x)[0][0]
win = pipe.predict_proba(x)[0][1]
print("Probability of " + abbreviation[batting_team] + " winning is " + str(win*100))
print("Probability of " + abbreviation[batting_team] + " losing is " + str(lose*100))
```
